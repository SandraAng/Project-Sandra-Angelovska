## Scraping articles and data collection

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#scraping sports news
```

```python
url = 'https://www.bbc.com/sport'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

sport_articles = soup.find_all('a', class_="ssrcss-zmz0hi-PromoLink exn3ah91")
sport_data = []
for sport_article in sport_articles:
    sport_article_text=sport_article.text.strip()
    sport_data.append((sport_article_text, 'sport'))
print(sport_data)
 
            
```

```python
#scraping climate change news
```

```python
url_1 = 'https://www.bbc.com/news/science-environment-56837908'
response = requests.get(url_1)

new_soup = BeautifulSoup(response.text, 'html.parser')

climate_articles = new_soup.find_all('a', class_="qa-heading-link lx-stream-post__header-link")

climate_data = []
for climate_article in climate_articles:
    climate_title_text = climate_article.text.strip()
    climate_data.append((climate_title_text, 'climate'))
print(climate_data)
    


```

```python
#the BBC website was not publishing enough new articles for climate change so to keep the amount of news articles for all categoreis balances,I decided to scrape an additional website for climate change news
```

```python
url_3 = 'https://www.climatechangenews.com/'
response = requests.get(url_3)
soup_3 = BeautifulSoup(response.text, 'html.parser')

additional_climate_articles = soup_3.find_all('h2')
additonal_climate_data = []

for additional_climate_article in additional_climate_articles:
    additional_climate_text = additional_climate_article.text.strip()
    additonal_climate_data.append((additional_climate_text, 'climate'))
print(additonal_climate_data)
    

   

   
```

```python
# scraping war in ukraine news
```

```python
url_2= 'https://www.bbc.com/news/world-60525350'
response = requests.get(url_2)

brand_new_soup = BeautifulSoup(response.text, 'html.parser')

war_articles = brand_new_soup.find_all('a', class_="gs-c-promo-heading gs-o-faux-block-link__overlay-link gel-pica-bold nw-o-link-split__anchor")
war_data = []

for war_article in war_articles:
    war_title_text = war_article.text.strip()
    war_data.append((war_title_text, 'war'))
print(war_data)
    

    
```

## Processing the text

```python
import nltk
from wordcloud import WordCloud
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import re
```

```python
df = pd.read_excel('C:\\Users\\Saso\\IT semester 2\\articles.xlsx')
```

```python
category_counts = df['Category'].value_counts()
category_counts.plot(kind='bar', color=["blue","green", "red"])
plt.xlabel("Categories of articles")
plt.ylabel("Number of articles")
plt.title("Articles per Category")
```

```python
def remove_tags(text):
    clean = re.compile('<.*?>')
    text = re.sub(clean, '', text)
    
    clean = re.compile('[^a-zA-Z0-9\s]')
    text = re.sub(clean, '', text)
    
    return text

df['Title'] = df['Title'].apply(remove_tags)

df_processed = pd.DataFrame({
    'Title': df['Title'], 
    'Category': df['Category'],  
    'Label': df['Label'] 
})

print(df_processed)
```

```python
def generate_word_cloud(text, title):
    stopwords_set = set(stopwords.words('english'))
    
   
    text = text.lower()
    words = nltk.word_tokenize(text)
    words = [word for word in words if word not in stopwords_set]
    cleaned_text = ' '.join(words)
    
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(cleaned_text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title)
    plt.axis('off')
    plt.show()

for category in category_mapping.keys():
    category_text = ' '.join(df_processed[df_processed['Category'] == category]['Title'])
    generate_word_cloud(category_text, f'Most Common Words for {category.capitalize()}')
```

## Feature extraction with Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
y = df_processed['Category']  
x_text = df_processed['Title'] 

cv = CountVectorizer(max_features=500)
x = cv.fit_transform(x_text).toarray()

print("x.shape = ", x.shape)
print("y.shape = ", y.shape)

print("Vocabulary: ")
print(cv.get_feature_names_out())
```

## Spliting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("x_train.shape =", x_train.shape)
print("x_test.shape =", x_test.shape)
print("y_train.shape =", y_train.shape)
print("y_test.shape =", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
#Create a Logistic Regression model
logistic_regression_model = LogisticRegression(max_iter=1000, random_state=42)

# Train the model on the training data
logistic_regression_model.fit(x_train, y_train)

# Predict categories for the test data
y_pred = logistic_regression_model.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest classifier
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training data
random_forest_model.fit(x_train, y_train)

# Predict categories for the test data
y_pred_rf = random_forest_model.predict(x_test)

# Evaluate the Random Forest model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
classification_rep_rf = classification_report(y_test, y_pred_rf)

print("Random Forest Classifier Results:")
print("Accuracy:", accuracy_rf)
print("Confusion Matrix:")
print(conf_matrix_rf)
print("Classification Report:")
print(classification_rep_rf)
```
